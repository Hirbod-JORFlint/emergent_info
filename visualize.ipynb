```python
# Global Baselined Probing (GBP)
def baselined(lst):
    return [lst[i] - lst[0] for i in range(1,13)]

# Global Conditional Probing (GCP)
def conditional(lst, lst_0i):
    return [lst_0i[i] - lst[0] for i in range(0,12)]

# Local Baselined Probing (LBP)
def baselined_ig(lst):
    return [lst[i] - lst[i-1] for i in range(1,13)]

# Local Conditional Probing (LCP)
def conditional_ig(lst, lst_pr):
    return [lst_pr[i] - lst[i] for i in range(0,12)]
```

```python
# Emergent Information (EMI)
def contributions(lst, lst_pr):
    layers = [lst_pr[i] - lst[i] for i in range(0,12)]
    denom = 0
    for i in range(0,12):
        if layers[i] > 0:
            denom += layers[i]
    return [100*layers[i] / denom  if layers[i] / denom > 0 else 0 for i in range(0,12)]


# Early Contributions for layer 1, 1+2, 1+2+3
def early_gains(lst, lst_pr):
    c = contributions(lst, lst_pr)
    return(c[0], c[0]+c[1], c[0]+c[1]+c[2])
    

print('cs')
print(early_gains(cs_norm, cs_norm_pr))
print(early_gains(cs_easy, cs_easy_pr))
print(early_gains(cs_hard, cs_hard_pr))

print('de')
print(early_gains(de_norm, de_norm_pr))
print(early_gains(de_easy, de_easy_pr))
print(early_gains(de_hard, de_hard_pr))

print('en')
print(early_gains(en_norm, en_norm_pr))
print(early_gains(en_easy, en_easy_pr))
print(early_gains(en_hard, en_hard_pr))

print('fi')
print(early_gains(fi_norm, fi_norm_pr))
print(early_gains(fi_easy, fi_easy_pr))
print(early_gains(fi_hard, fi_hard_pr))

print('he')
print(early_gains(he_norm, he_norm_pr))
print(early_gains(he_easy, he_easy_pr))
print(early_gains(he_hard, he_hard_pr))

print('sv')
print(early_gains(sv_norm, sv_norm_pr))
print(early_gains(sv_easy, sv_easy_pr))
print(early_gains(sv_hard, sv_hard_pr))

print('tr')
print(early_gains(tr_norm, tr_norm_pr))
print(early_gains(tr_easy, tr_easy_pr))
print(early_gains(tr_hard, tr_hard_pr))

#get_plot2(test(en_norm, en_norm_pr), test(en_easy, en_easy_pr), test(en_hard, en_hard_pr))

```

```python
# EMI-BL
def mi_bl(lst):
    layers = [lst[i] - lst[i-1] for i in range(1,13)]
    denom = 0
    for i in range(0,12):
        if layers[i] > 0:
            denom += layers[i]
    return [layers[i] / denom  if layers[i] / denom > 0 else 0 for i in range(0,12)]


def cont_vs_bl(lst, lst_pr):
    #print(contributions(lst, lst_pr), mi_bl(lst))
    l = [abs(a-100*b) for a,b in zip(contributions(lst, lst_pr), mi_bl(lst))]
    l = l[0:2]
    return sum(l) / len(l)


# What's the avg difference between with/without conditioning? 
# POS
avg_all_pos = [cont_vs_bl(en_easy, en_easy_pr), cont_vs_bl(de_easy, de_easy_pr), cont_vs_bl(cs_easy, cs_easy_pr), cont_vs_bl(fi_easy, fi_easy_pr), cont_vs_bl(tr_easy, tr_easy_pr), 
cont_vs_bl(sv_easy, sv_easy_pr), cont_vs_bl(he_easy, he_easy_pr), cont_vs_bl(en_hard, en_hard_pr), cont_vs_bl(he_hard, he_hard_pr), cont_vs_bl(de_hard, de_hard_pr), 
cont_vs_bl(cs_hard, cs_hard_pr), cont_vs_bl(tr_hard, tr_hard_pr), cont_vs_bl(fi_hard, fi_hard_pr)]
print(sum(avg_all_pos) / len(avg_all_pos))

# Ancestors
avg_all_anc = [cont_vs_bl(parents_de, parents_de_pr), cont_vs_bl(parents_fi, parents_fi_pr), cont_vs_bl(parents_he, parents_he_pr), cont_vs_bl(parents_en, parents_en_pr), 
cont_vs_bl(parents_tr, parents_tr_pr), cont_vs_bl(parents_sv, parents_sv_pr), cont_vs_bl(parents_cs, parents_cs_pr), cont_vs_bl(grandparents_de, grandparents_de_pr), 
cont_vs_bl(grandparents_fi, grandparents_fi_pr), cont_vs_bl(grandparents_he, grandparents_he_pr), cont_vs_bl(grandparents_en, grandparents_en_pr), 
cont_vs_bl(grandparents_tr, grandparents_tr_pr), cont_vs_bl(grandparents_sv, grandparents_sv_pr), cont_vs_bl(grandparents_cs, grandparents_cs_pr)]
print(sum(avg_all_anc) / len(avg_all_anc))
```
